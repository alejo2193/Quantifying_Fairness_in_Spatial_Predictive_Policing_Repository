```python
import pickle as pkl
import os
import datetime

#Cargando los datos
os.chdir("drive/Shareddrives/FAIRNESS/Colab/Notebooks_for_topic/EXP0/")
```

##<center> <font color='green'> **0. Setup variables**

```python
from global_vars import f_inicial,f_final,f_final_train,f_final_test,f_final_val
from global_vars import dir_sims, dir_split
prefix_data="Data_"
```

```python
for i in os.listdir(dir_sims):
    if prefix_data in i:
        timed_points_NM = pkl.load(open(os.path.join(dir_sims,i), "rb")) ## cargar datos

        data_train = timed_points_NM[(timed_points_NM.times_datetime()>=f_inicial) & ((timed_points_NM.times_datetime()<f_final_train))]
        data_test = timed_points_NM[(timed_points_NM.times_datetime()>=f_final_train) & ((timed_points_NM.times_datetime()<f_final_test))]
        data_val = timed_points_NM[(timed_points_NM.times_datetime()>=f_final_test) & ((timed_points_NM.times_datetime()<f_final_val))]

        pkl.dump(data_train, open(os.path.join(dir_split,"Train_"+i), "wb"))
        pkl.dump(data_test, open(os.path.join(dir_split,"Test_"+i), "wb"))
        pkl.dump(data_val, open(os.path.join(dir_split,"Val_"+i), "wb"))
```
